```python
import os
import pandas as pd
import numpy as np
from difflib import SequenceMatcher

def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()

def swap(col, guide):
    
    for i in range(len(col)):
        temp = 0
        index = -1
    
        for j in range(len(guide)):
            if similar(col[i],guide[j]) > temp:
                index = j
                temp = similar(col[i],guide[j])
            else:
                continue
        
        if temp >= 0.6:
            col[i] = guide[index]
        else:
            continue

    return col

addresses = ['./resources/tracks-1921-2020.csv',
             './resources/top-tracks-2019.csv',
             './resources/top-tracks-2018.csv',
             './resources/top-tracks-2017.csv',
             './resources/top-tracks-2010-to-2019.csv']

df1 = pd.read_csv(addresses[0])
df2 = pd.read_csv(addresses[1])
df3 = pd.read_csv(addresses[2])
df4 = pd.read_csv(addresses[3])
df5 = pd.read_csv(addresses[4])

col = []
guide = []

#adding strings for column name guidance list
for column_name in df1.columns:
    guide.append(column_name)

guide.append('genre')
print(guide)

#implementation with string similarity comparison for column names of df2
for column_name in df2.columns:
    col.append(column_name)

print('\ndf2-Before: ')
print(col)
col = swap(col, guide)
df2.columns = col

#manual change for uncovered cases
new_columns = df2.columns.values
new_columns[1] = 'name'
new_columns[2] = 'artists'
new_columns[4] = 'tempo'
new_columns[10] = 'duration_ms'
df2.columns  = new_columns
print('\ndf2-After: ')
print(df2.columns.values)

#implementation with string similarity comparison for column names of df3
col = []
for column_name in df3.columns:
    col.append(column_name)

print('\ndf3-Before: ')
print(col)
col = swap(col, guide)
df3.columns = col

print('\ndf3-After: ')
print(df3.columns.values)

#implementation with string similarity comparison for column names of df4
col = []
for column_name in df4.columns:
    col.append(column_name)

print('\ndf4-Before: ')
print(col)
col = swap(col, guide)
df4.columns = col

print('\ndf4-After: ')
print(df4.columns.values)

#implementation with string similarity comparison for column names of df5
col = []
for column_name in df5.columns:
    col.append(column_name)

print('\ndf5-Before: ')
print(col)
col = swap(col, guide)
df5.columns = col

#manual change for uncovered cases
new_columns = df5.columns.values
new_columns[1] = 'name'
new_columns[5] = 'tempo'
new_columns[7] = 'danceability'
new_columns[8] = 'loudness'
new_columns[11] = 'duration_ms'
new_columns[12] = 'acousticness'
new_columns[13] = 'speechiness'
new_columns[14] = 'popularity'
df5.columns  = new_columns
print('\ndf5-After: ')
print(df5.columns.values)
```
